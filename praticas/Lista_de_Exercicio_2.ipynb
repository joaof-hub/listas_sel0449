### 2) Resolução de escala de cinza (peso 3/10)

1. Agora vamos reduzir a resolução de níveis de cinza da imagem. Para isto, vamos usar uma função de nome ```transformaImg```, já disponibilizada a seguir. Comente o que cada linha da função está exercendo.

2. Carregue o arquivo de imagem ```angiography.bmp``` e mostre a imagem. Limite os valores máximo e mínimo do seu plot para ```vmin=0``` e ```vmax=255```.

3. Reduza a escala de níveis de cinza da imagem ```angiography.tif``` de 256 níveis para a resoluções abaixo. Mostre todas as imagens usando janelas de mesmo tamanho, assim como foi feito/explicado no exercício anterior. Limite os valores máximo e mínimo do seu plot para ```vmin=0``` e ```vmax=255```.

  - 128
  - 64
  - 16
  - 4
  - 2
  

4. Comente os resultados obtidos.

5. Explique:
  * Qual o motivo da transformação da variável img para ```float32``` no início da função dada?
  * Qual o motivo da linha ```res = 255*(res - res.min())/(res.max() - res.min())``` antes do retorno para ```uint8```?

````
#1
def transformaImg(img, grayLevel):
  '''
  Descrição: Transforma uma imagem para uma nova escala de cinza.

  Entrada:
    - img: Imagem de entrada (uint8).
    - grayLevel: Nova escala de níveis de cinza desejada (uint8).

  Saída:
    - res: Imagem com nova escala de cinza (uint8)
  '''

  img_in = img.copy().astype('float32')

  res = np.round( img_in * (grayLevel - 1) / 255.)

  res = 255*(res - res.min())/(res.max() - res.min())

  res = res.astype('uint8')

  return res

# Seu código começa aqui
#lê a imagem inalterada
img2= cv.imread("angiography.tif", cv.IMREAD_UNCHANGED)

#plota a figura no canal gray , com vmin=0 e vmax=255
plt.figure()
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.show()


levels=[128,64,16,4,2] #cria um array com as novas amplitudes/resoluções do nível
# de cinza
images=[img2]#cria  um array com as imagens a serem plotadas

# lê as amplitudes de cinza no array e aplica cada uma delas na imagem original
# usando um loop e armazena no array de imagens
for level in levels:
  images.append(transformaImg(img2, level))

#plota todas as imagens no array usando um loop
for image in images:
  plt.figure()
  plt.imshow(image, cmap='gray', vmin=0, vmax=255)
  plt.show()
# como a imagem teve sua amplitude de níveis de cinza reduzida pela função
#"transformaImg" os pixels com valores próximos de intensidade foram admitidos
# como tendo o mesmo valor no programa
#Isso faz com que ao  plotar novamente as imagens a resolução delas seja reduzida
# já que pixels com intensidades similares passaram a ser representados com o
#mesmo valor

# A linha img_in = img.copy().astype('float32') é necessária pois a imagem
#original precisa registrar corretamente as casas decimais na linha
#  res = np.round( img_in * (grayLevel - 1) / 255.) na expressão dentro do
# parentesis , o que não aconteceria se a imagem fosse do tipo int ocasionando
# o truncamento prematuro antes do arredondamento

# A linha "res = 255*(res - res.min())/(res.max() - res.min())"  Faz com que a
#amplitude de intensidade seja novamente registrada entre 0 e 255, porém mantendo
# a proporção de valores da imagem reduzida.Dessa forma ,embora a imagem seja
# registrada com valores de pixels correspondentes nominalmente a uma faixa de
#0 a 255 a resolução real corresponde a uma faixa de menor amplitude , o que faz
# com que a imagem em escala cinza possua menor nitidez mesmo no intervalo de
#0 a 255




# Seu código termina aqui
````
